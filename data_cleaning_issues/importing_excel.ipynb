```python
# import pandas
import pandas as pd
```

```python
pd.options.display.float_format = '{:.0f}'.format
pd.set_option('display.width', 60)
pd.set_option('display.max_columns', None)
```

```python
# import the land temperature data
percapitaGDP = pd.read_excel("data/GDPpercapita22b.xlsx",
   sheet_name="OECD.Stat export",
   skiprows=4,
   skipfooter=1,
   usecols="A,C:W")
```

```python
# Use head to view the first few rows and shape to get the number of rows and columns:
percapitaGDP.head()
```

```python
percapitaGDP.shape
```

```python
# Use the info method of the DataFrame to view data types and the non-null count
# NOTE all columns have the object data type
percapitaGDP.info()
```

```python
# Give an appropriate name to the metropolitan area column
percapitaGDP.rename(columns={'Year':'metro'}, inplace=True)
```

```python
#Test for leading spaces with startswith(' ') and then use any to establish whether there are one or more occasions when the first character is blank.
percapitaGDP.metro.str.startswith(' ').any()
```

```python
# Use endswith(' ') to examine trailing spaces.
percapitaGDP.metro.str.endswith(' ').any()
```

```python
#Use stripmethod to remove all leading spaces.
percapitaGDP.metro = percapitaGDP.metro.str.strip()
```

```python
# Test
percapitaGDP.metro.str.startswith(' ').any()
```

```python
# convert the data columns to numeric
for col in percapitaGDP.columns[1:]:
  percapitaGDP[col] = pd.to_numeric(percapitaGDP[col],
    errors='coerce')
  percapitaGDP.rename(columns={col:'pcGDP'+col},
    inplace=True)
```

```python
# Remove rows where all of the per capita GDP values are missing.
percapitaGDP.dropna(subset=percapitaGDP.columns[1:], how="all", inplace=True)
```
